# seperating the data and the labels

```python
aa = a.drop(['R'],axis = 1)
bb = a['R']
```

```python
print(aa)
print(bb)
```


# training and test the data 

```python
aa_train, aa_test, bb_train, bb_test = train_test_split(aa,bb,test_size = 0.1 , stratify = bb, random_state = 1 )
```

```python
print(aa.shape, aa_train.shape , aa_test.shape )
```

```python
print(aa_train)
print(bb_train)
```

# model training --> logistic regression

```python
model = LogisticRegression()
```

# training the logisticregression model with the training data 

```python
model.fit(aa_train , bb_train)
```

# model evalutaion 

```python
#accuracy on the training the data 
```

```python
aa1 = model.predict(aa_train)
aa2 = accuracy_score(aa1, bb_train)
```

```python
print('accuracy on the training the model  = ', aa2)
```

```python
#accuracy on the testing the data 
```

```python
aa3 = model.predict(aa_test)
aa4 = accuracy_score(aa3, bb_test)
```

```python
print('accuracy on teh testing the model = ',aa4 )
```
