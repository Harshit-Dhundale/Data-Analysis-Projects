```python
from sklearn.preprocessing import LabelEncoder  # for converting the string value in the integer for applying the linearregression 
```

```python
le_Gender = LabelEncoder()
le_lungcancer = LabelEncoder()
```

```python
a['sex'] = le_Gender.fit_transform(a['GENDER'])
a['lungcancer'] = le_lungcancer.fit_transform(a['LUNG_CANCER'])
a
```

```python
"""
sex 
male = 1
female = 0

lungcancer 
yes = 1
no = 0


any disease
yes = 2
no = 1

"""
```

```python
from sklearn import linear_model
```

```python
r = linear_model.LinearRegression()
r.fit(a[['sex','AGE','SMOKING','ALCOHOL CONSUMING','CHEST PAIN']],a.lungcancer)
```
